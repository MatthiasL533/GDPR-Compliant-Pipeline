```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Various Sector": [
        "Media, Telecoms and Broadcasting", "Industry and Commerce", "Transportation and Energy", 
        "Finance, Insurance and Consulting", "Employment", "Public Sector and Education", 
        "Accommodation and Hospitality", "Health Care", "Real Estate", 
        "Individuals and Private Associations", "Not assigned"
    ],
    "Sum of Fines (€)": [
        3313691366, 913318701, 167085710, 62236758, 59598777, 
        27827063, 22564648, 17071309, 2601831, 1890856, 1814708
    ],
    "Number of Fines": [
        295, 455, 115, 221, 140, 241, 70, 205, 60, 290, 132
    ]
}

# Convert Sum of Fines to millions
data["Sum of Fines (€)"] = [x / 1_000_000 for x in data["Sum of Fines (€)"]]

# Create DataFrame
df = pd.DataFrame(data)

# Plot
fig, ax1 = plt.subplots(figsize=(14, 8))

# Bar plot for sum of fines
color = 'dodgerblue'
ax1.set_xlabel('Various Sector', fontsize=14)
ax1.set_ylabel('Sum of Fines (€ Million)', color=color, fontsize=14)
bars = ax1.bar(df['Various Sector'], df['Sum of Fines (€)'], color=color, alpha=0.7, label='Sum of Fines (€ Million)')
ax1.tick_params(axis='y', labelcolor=color, labelsize=12)
ax1.tick_params(axis='x', labelsize=12)
plt.xticks(rotation=45, ha='right')

# Adding labels to the bars
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:,.2f}', va='bottom', ha='center', fontsize=10, color='black')

# Line plot for number of fines
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Number of Fines', color=color, fontsize=14)
line = ax2.plot(df['Various Sector'], df['Number of Fines'], color=color, marker='o', label='Number of Fines')
ax2.tick_params(axis='y', labelcolor=color, labelsize=12)

# Adding labels to the points
for i, txt in enumerate(df['Number of Fines']):
    ax2.annotate(txt, (df['Various Sector'][i], df['Number of Fines'][i]), textcoords="offset points", xytext=(0,5), ha='center', fontsize=10, color='black')

# Title and Layout
fig.suptitle('Sum and Number of Fines by Various Sectors for GDPR Violations', fontsize=16, fontweight='bold')
fig.tight_layout(rect=[0, 0, 1, 0.96])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data from your table
datasets = ['df1', 'df2', 'df3', 'df4', 'df5', 'df6', 'df7', 'df8']
p29_scores = [0, 0, 0, 0, 0, 0.8358, 0.8367, 0.8321]
min_k_values = [1, 1, 1, 1/9, 1/26, 1/221, 1/221, 1/221]
min_l_values = [0, 0, 0, 0, 0, 0.0678, 0.0678, 0.0678]
max_t_values = [0.9193, 0.9193, 0.9193, 0.6942, 0.6412, 0.4565, 0.4565, 0.4565]

# Plotting
plt.figure(figsize=(10, 6))

# Plotting P29 scores as bars
bars = plt.bar(datasets, p29_scores, color='skyblue', label='P29 Score')

# Plotting minimum k values
dots_k = plt.scatter(datasets, min_k_values, color='hotpink', marker='o', label='1/Minimum k')

# Plotting minimum normalized entropy l values
dots_l = plt.scatter(datasets, min_l_values, color='seagreen', marker='s', label='Minimum normalized entropy l')

# Plotting maximum t values
dots_t = plt.scatter(datasets, max_t_values, color='darkorange', marker='^', label='Maximum t')

# Adding connecting lines between dots across datasets
for i in range(len(datasets) - 1):
    
    plt.plot([datasets[i], datasets[i+1]], [min_k_values[i], min_k_values[i+1]], color='hotpink', linestyle='--', linewidth=1)
    plt.plot([datasets[i], datasets[i+1]], [min_l_values[i], min_l_values[i+1]], color='seagreen', linestyle='--', linewidth=1)
    plt.plot([datasets[i], datasets[i+1]], [max_t_values[i], max_t_values[i+1]], color='darkorange', linestyle='--', linewidth=1)

# Adding labels and title
plt.xlabel('Dataset Number')
plt.ylabel('Metrics Values')
plt.title('GDPR-Compliant Privacy Metrics')
plt.xticks(rotation=30)

# Adding numerical values next to each point
for dataset, p29, k, l, t in zip(datasets, p29_scores, min_k_values, min_l_values, max_t_values):
    plt.text(dataset, p29, f'{p29:.4f}', ha='center', va='bottom', fontsize=8, color='skyblue')
    plt.text(dataset, k, f'{k:.4f}', ha='center', va='bottom', fontsize=8, color='hotpink')
    plt.text(dataset, l, f'{l:.4f}', ha='center', va='bottom', fontsize=8, color='seagreen')
    plt.text(dataset, t, f'{t:.4f}', ha='center', va='bottom', fontsize=8, color='darkorange')

# Adjust legend position to avoid covering the plot
plt.legend(loc='center left')

# Adding grey grid
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Adjust layout to prevent clipping of labels
plt.tight_layout()
plt.savefig('privacy_metrics_comparison.png')
# Show plot
plt.show()


```
