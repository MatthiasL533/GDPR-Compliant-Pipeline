```python
def drop_identifier(df):
    """
    Interactively drop specified columns from a DataFrame.

    Parameters:
    - df (pd.DataFrame): The DataFrame from which columns will be dropped.

    Returns:
    - pd.DataFrame: A new DataFrame with specified columns removed.
    """
    # List all columns in the DataFrame and ask the user to input the column names to drop
    print("Current columns in DataFrame:", df.columns.tolist())
    columns_to_drop = input("Enter the column names to drop (separated by comma): ").split(',')
    
    # Strip spaces and handle case sensitivity if needed
    columns_to_drop = [col.strip() for col in columns_to_drop]

    # Check if the columns exist in the DataFrame
    existing_columns_to_drop = [col for col in columns_to_drop if col in df.columns]
    missing_columns = [col for col in columns_to_drop if col not in df.columns]

    if missing_columns:
        print(f"Warning: The following columns were not found in the DataFrame and will be ignored: {missing_columns}")

    # Drop the columns that exist in the DataFrame
    df = df.drop(columns=existing_columns_to_drop)
    
    print(f"The following columns were successfully dropped: {existing_columns_to_drop}")
    return df

# Example usage:
# df = drop_identifier(df)
drop_identifier(df)
```

```python
#check the features with missing values and decide whether drop them or not

def drop_columns_with_missing_values(df):
    # Calculating percentage of missing values for each column
    missing_percentages = (df.isnull().sum() / len(df)) * 100
    
    # Printing columns with missing values and their percentages
    print("Columns with missing values:")
    print(missing_percentages[missing_percentages > 0])
    
    # Selecting columns to drop
    columns_to_drop = input("Enter column names to drop (separated by commas), or 'none' to keep all: ")
    
    # Dropping selected columns
    if columns_to_drop.lower() != 'none':
        columns_to_drop = [col.strip() for col in columns_to_drop.split(',')]
        df.drop(columns=columns_to_drop, inplace=True)
        print("Columns dropped successfully.")
    else:
        print("No columns were dropped.")

drop_columns_with_missing_values(df)The,
```

```python
# Choose the qusi-identifier and Data Generalisation(generalise the data through mapping)

def identify_quasi_identifiers_with_distinct_values(df):
    """
    Identifies quasi-identifiers based on user-selected columns and shows their distinct values.
    """
    # Select columns to identify as quasi-identifiers
    selected_columns = input("Enter column names to identify as quasi-identifiers (separated by commas): ").split(',')
    selected_columns = [col.strip() for col in selected_columns]
    
    # Verify that the selected columns exist in the DataFrame
    selected_columns = [col for col in selected_columns if col in df.columns]
    
    if not selected_columns:
        print("No valid columns selected.")
        return []
    
    # Loop through the selected columns to calculate the distinct values and percentages
    for column in selected_columns:
        distinct_values = df[column].value_counts()
        total_values = len(df[column])
        percentages = (distinct_values / total_values) * 100
        
        print(f"Distinct values and percentages for '{column}':")
        for value, percentage in zip(distinct_values.index, percentages):
            print(f"{value}: {percentage:.2f}%")
        print()
    
    return selected_columns

def map_values_and_output_percentages(df, selected_columns):
    """
    Maps values in the selected columns to other values based on user input,
    including the ability to map to newly created values or remap values after reviewing. Outputs updated percentages.
    """
    mapping_dict = {}

    # Allow users to remap values if needed
    continue_mapping = True
    while continue_mapping:
        for column in selected_columns:
            if column not in df.columns:
                print(f"Column '{column}' not found in the DataFrame.")
                continue

            # Initialize or clear the column mapping
            mapping_dict[column] = {}
            
            while True:
                # List available values for mapping in the current column
                distinct_values = df[column].value_counts().index.tolist()
                print(f"\nAvailable values in '{column}': {distinct_values}")
                
                # Prompt user to map values
                original_value = input(f"Enter a value from '{column}' to map, type 'new' to create a new value, or 'done' to finish: ").strip().lower()
                if original_value == 'done':
                    break
                if original_value == 'new':
                    new_value = input(f"Enter the name of the new value for column '{column}': ").strip().lower()
                    df[column] = df[column].astype(str)  # Ensure data type consistency
                    df.loc[~df[column].str.lower().isin([v.lower() for v in distinct_values]), column] = new_value
                    print(f"New value '{new_value}' has been added to available values.")
                    continue

                if original_value not in [value.lower() for value in distinct_values]:
                    print(f"'{original_value}' is not a valid value for column '{column}'. Try again.")
                    continue

                # Get the target value for mapping
                target_value = input(f"Enter the value to which '{original_value}' should be mapped, or type 'new' to create a new value: ").strip().lower()
                if target_value == 'new':
                    target_value = input(f"Enter the name of the new value to map '{original_value}' to: ").strip().lower()

                # Confirm mapping before proceeding
                confirm = input(f"Map '{original_value}' to '{target_value}'? Type 'yes' to confirm or 'no' to re-enter: ").strip().lower()
                if confirm == 'yes':
                    mapping_dict[column][original_value] = target_value
                else:
                    print("Mapping not confirmed. Please re-enter your mappings.")

            # Apply the mapping to the DataFrame
            df[column] = df[column].apply(lambda x: mapping_dict[column].get(x.lower(), x))

        # Output the updated percentages for each column
        print("\nUpdated distinct values and percentages for all mapped columns:")
        for column in selected_columns:
            updated_value_counts = df[column].value_counts()
            total_values = len(df[column])
            updated_percentages = (updated_value_counts / total_values) * 100
            print(f"{column}:")
            for value, percentage in zip(updated_value_counts.index, updated_percentages):
                print(f"{value}: {percentage:.2f}%")
        
        # Ask if the user wants to continue mapping
        continue_mapping = input("\nDo you want to remap any columns? Type 'yes' to continue or 'no' to finish: ").strip().lower() == 'yes'


selected_columns = identify_quasi_identifiers_with_distinct_values(df)

# Perform mapping actions and output the updated percentages
if selected_columns:
    map_values_and_output_percentages(df, selected_columns)
```
